```python
from xpsi.tools import phase_interpolator

# Interpolating and summing spot1 (in signal.signal[0]) and spot2 (in signal.signal[1])
# with their respective signal phase shifts.

# SPOT 1 + SPOT 2
signal_signal = phase_interpolator(main.XTI.data.phases, main.XTI.signal.phases[0], main.XTI.signal.signals[0], main.XTI.signal.shifts[0]) + \
                phase_interpolator(main.XTI.data.phases, main.XTI.signal.phases[0], main.XTI.signal.signals[1], main.XTI.signal.shifts[1])

XpsiPlot.plot_2d_pulse(signal_signal,
                       x=main.XTI.data.phases,
                       y=main.XTI.data.channels,
                       rotations=2,
                       ylabel=r'Channels',
                       cbar_label='Counts (normalized by maximum)',
                       yticks=[30,50,100,200],
                       normalized=True,
                      )
```
